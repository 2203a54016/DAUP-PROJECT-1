```python
# Scatter plots- with and without outliers,  skewness,Â kurtosis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

# Load the dataset
csv_path = "/content/spam email classification numerical dataset.zip"  # Update path if necessary
df = pd.read_csv(csv_path)

# Select a feature for analysis
feature = "capital_run_length_total"
print(feature)

# Compute skewness and kurtosis
skewness = skew(df[feature])
kurt = kurtosis(df[feature])

print(f"Skewness of {feature}: {skewness:.2f}")
print(f"Kurtosis of {feature}: {kurt:.2f}")

# Scatter plot with outliers
plt.figure(figsize=(12, 5))
sns.scatterplot(x=df.index, y=df[feature], color='red')
plt.title(f"Scatter Plot of {feature} (With Outliers)")
plt.xlabel("Index")
plt.ylabel(feature)
plt.show()

# Removing outliers using IQR
Q1 = df[feature].quantile(0.25)
Q3 = df[feature].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_no_outliers = df[(df[feature] >= lower_bound) & (df[feature] <= upper_bound)]

# Scatter plot without outliers
plt.figure(figsize=(12, 5))
sns.scatterplot(x=df_no_outliers.index, y=df_no_outliers[feature], color='blue')
plt.title(f"Scatter Plot of {feature} (Without Outliers)")
plt.xlabel("Index")
plt.ylabel(feature)
plt.show()
```

```python
# Histogram plots- with and without outliers,  skewness,Â kurtosis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

# Load the dataset
csv_path = "/content/spam email classification numerical dataset.zip"
df = pd.read_csv(csv_path)

# Select a numerical column for analysis
column = "capital_run_length_total"

# Compute skewness and kurtosis
original_skewness = skew(df[column])
original_kurtosis = kurtosis(df[column])

# Identify outliers using the IQR method
Q1 = df[column].quantile(0.25)
Q3 = df[column].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter data without outliers
df_no_outliers = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Compute skewness and kurtosis without outliers
filtered_skewness = skew(df_no_outliers[column])
filtered_kurtosis = kurtosis(df_no_outliers[column])

# Plot histograms
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Original distribution with outliers
sns.histplot(df[column], bins=30, kde=True, ax=axes[0], color='red')
axes[0].set_title(f"With Outliers\nSkewness: {original_skewness:.2f}, Kurtosis: {original_kurtosis:.2f}")
axes[0].set_xlabel(column)

# Distribution without outliers
sns.histplot(df_no_outliers[column], bins=30, kde=True, ax=axes[1], color='blue')
axes[1].set_title(f"Without Outliers\nSkewness: {filtered_skewness:.2f}, Kurtosis: {filtered_kurtosis:.2f}")
axes[1].set_xlabel(column)

plt.tight_layout()
plt.show()
```
